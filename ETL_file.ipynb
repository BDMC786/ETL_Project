```python
import pandas as pd
base_url = "https://www.calendar-12.com/moon_phases/"
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]
```

```python
moons = []
for year in years:
    tables = pd.read_html(f'{base_url}{year}')
    df = tables[0]
    full_df=df['Unnamed: 2_level_0']
    full_df = full_df.dropna(how='any')    
    for x in full_df["Full Moon"]:
        z = x.split(",")[0]
        if z[0:4] == "Jan.":
            moon = z.replace("Jan.", "January")
            moon = f'{moon} {year}'
            moons.append(moon)
        elif z[0:4] == "Feb.":
            moon = z.replace("Feb.", "February")
            moon = f'{moon} {year}'
            moons.append(moon)
        elif z[0:4] == "Mar.":
            moon = z.replace("Mar.", "March")
            moon = f'{moon} {year}'
            moons.append(moon)
        elif z[0:4] == "Apr.":
            moon = z.replace("Apr.", "April")
            moon = f'{moon} {year}'
            moons.append(moon)
        elif z[0:4] == "Aug.":
            moon = z.replace("Aug.", "August")
            moon = f'{moon} {year}'
            moons.append(moon)
        elif z[0:4] == "Sept":
            moon = z.replace("Sept.", "September")
            moon = f'{moon} {year}'
            moons.append(moon)
        elif z[0:4] == "Oct.":
            moon = z.replace("Oct.", "October")
            moon = f'{moon} {year}'
            moons.append(moon)
        elif z[0:4] == "Nov.":
            moon = z.replace("Nov.", "November")
            moon = f'{moon} {year}'
            moons.append(moon)
        elif z[0:4] == "Dec.":
            moon = z.replace("Dec.", "December")
            moon = f'{moon} {year}'
            moons.append(moon)
        else:
            z = f'{z} {year}'
            moons.append(z)
moons
```

```python
len(moons)
```

```python
col = []
for moon in moons:
    col.append("Full Moon")
```

```python
len(col)
```

```python
moon_df = pd.DataFrame({"Date":moons, "Moon":col})
moon_df.head()
```

Done with Extracting and Transforming Data source 1

```python
raw_births_df = pd.read_csv("births.csv")
raw_births_df.head()
```

```python
date_col = []
for x in raw_births_df["date_of_month"]:
    date_col.append(x)

birth_col = []
for x in raw_births_df["births"]:
    birth_col.append(x)


year_col = []
for row in raw_births_df["year"]:
    year_col.append(row)


month_col = []
for row in raw_births_df["month"]:
    if row == 1:
        month_col.append("January")
    elif row == 2:
        month_col.append("February")
    elif row == 3:
        month_col.append("March")
    elif row == 4:
        month_col.append("April")
    elif row == 5:
        month_col.append("May")
    elif row == 6:
        month_col.append("June")
    elif row == 7:
        month_col.append("July")
    elif row == 8:
        month_col.append("August")
    elif row == 9:
        month_col.append("September")
    elif row == 10:
        month_col.append("October")
    elif row == 11:
        month_col.append("November")
    elif row == 12:
        month_col.append("December")    
```

```python
full_date = []

for i in range (0,len(date_col)):
    month = month_col[i]
    day = date_col[i]
    year = year_col[i]
    full_date.append(f'{month} {day} {year}')
len(full_date)
```

```python
births_df = pd.DataFrame({"Date":full_date, "Births":birth_col})
births_df.head()
```

```python
# births_df
# moon_df
# Merge 2 Dtaframes
merged_df = pd.merge(births_df, moon_df, how="left", on=["Date", "Date"])
merged_df.head()
```

```python
merged_df["Births"].sum()
```

```python
merged_df.Moon = merged_df.Moon.fillna(" ")
merged_df.head()
```

```python
revised = []
for x in merged_df["Moon"]:
    if x == "Full Moon":
        revised.append("Full")
    else:
        revised.append("Not Full")
revised
```

```python
merged_df["Moon"] = revised
merged_df.head()
```

```python
merged_df.to_csv("lunar_phases_vs_births.csv")
```
